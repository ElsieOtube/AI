```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from nltk.corpus import stopwords
stop = stopwords.words('english')

from sklearn.metrics.pairwise import cosine_similarity

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/content/drive/MyDrive/titles.csv')
print(df.shape)
df.head(10)
```

```python
df = df.sample(50)

#reset index
df = df.reset_index(drop=True)
df.head()
```

```python
# cleaning
df['ctitle'] = df['title'].str.replace('[^\w\s]','')
df['ctitle'] = df['ctitle'].str.replace('\d+', '')
df['ctitle'] = df['ctitle'].str.lower()

df['ctitle'] = df['ctitle'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

df.head()
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
features = model.encode(df['ctitle'])
df['features'] = list(features)
print(df['ctitle'].iloc[0])
#print(df['features'].iloc[0])
print(df['features'].iloc[0].shape)
```

```python
df
```

```python
def search(query):
  query_feature = model.encode([query])
  df['similarity'] = df['features'].apply( lambda x: cosine_similarity([x], query_feature) [0] [0])
  sorted_df = df.sort_values(by = 'similarity', ascending = False)
  return sorted_df.head(10)
```

```python
search('support tucker machine')
```

```python
def results(query):
  query_feature = model.encode([query])
  df['similarity'] = df['features'].apply( lambda x: cosine_similarity([x], query_feature) [0] [0])
  sorted_df = df.sort_values(by = 'similarity', ascending = False)
  res = sorted_df['title'].head(10)
  return list(res)
```

```python
results('support tucker machine')
```

```python
import gradio as gr

ui = gr.Interface(fn=results, inputs="text", outputs="text")
ui.launch(share = True)
```
